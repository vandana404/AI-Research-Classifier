```python
import gradio as gr
from transformers import pipeline

classifier = pipeline(
    "zero-shot-classification",
    model="facebook/bart-large-mnli"
)

DEFAULT_LABELS = [
    "Programming Language",
    "Framework",
    "AI Technique",
    "Model Architecture",
    "Medical Concept",
    "Research Area",
    "Database",
    "Cloud Platform",
    "Deployment Tool"
]

def classify_text(text):
    result = classifier(
        text,
        candidate_labels=DEFAULT_LABELS
    )

    return {
        label: score
        for label, score in zip(
            result["labels"],
            result["scores"]
        )
    }

interface = gr.Interface(
    fn=classify_text,
    inputs=gr.Textbox(
        lines=2,
        placeholder="Enter keyword or text..."
    ),
    outputs=gr.Label(num_top_classes=5),
    title="AI  Classifier",
    description="Classify research concepts automatically."
)

interface.launch()
```
